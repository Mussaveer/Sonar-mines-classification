```python
for i in Df.columns:
    if not pd.api.types.is_numeric_dtype(i):
        Df[i] = Df[i].astype('category')
        Df[i] = Df[i].cat.codes
```

```python
from scipy.stats import zscore
zscore = abs(zscore(Df))
print(Df.shape)
Df_new= Df.loc[(zscore<3).all(axis = 1)]
print(Df_new.shape)
```

```python
corr = Df_new.corr()
fig, ax = plt.subplots(figsize= (20,20))
sns.heatmap(corr,
           linewidth = 0.5,
           annot = False,
           fmt= '.2f');
```

* it seems like some of the features are strongly correlated

## Building the model

```python
X = Df_new.iloc[:, 0:-1]
y = Df_new.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
X_new = sd.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X_new,
                                                    y,
                                                    test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
models = {'RFC':RandomForestClassifier(),
         'LR' : LogisticRegression(),
         'KNN' : KNeighborsClassifier()}
model_score = {}
np.random.seed = 42
def fit_and_score(model, X_train, y_train, X_test, y_test):
    for name, model in models.items():
            model.fit(X_train, y_train)
            model_score[name] = model.score(X_test, y_test)
    return model_score
model_score = fit_and_score(model = models,
                           X_train = X_train,
                           y_train = y_train,
                           X_test = X_test,
                           y_test = y_test)    
        
```

```python
model_score
```

```python
#lets visualize it 
model_score_df = pd.DataFrame(model_score, index = ['Accuracy'])
model_score_df.T.plot.bar();
```

```python
#lets try to improve model(Random Forest Classifier) by hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators' : [100],
'max_depth' : [5, 8, 15, 25],
'min_samples_split' : [2, 5, 10, 15],
'min_samples_leaf' : [1, 2, 5, 10] }

RFC_grid = GridSearchCV(RandomForestClassifier(),
                       param_grid = param_grid,
                       cv = 2,
                       verbose = True)
RFC_grid.fit(X_train, y_train)
```

```python
RFC_grid.best_params_
```

```python
# lets evaluate the model
y_pred = RFC_grid.predict(X_test)
```

```python
RFC_grid.score(X_test, y_test)
```
